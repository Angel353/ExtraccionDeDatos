```python
# Importar las librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga desde un archivo .xlsx sin indice
df_pp = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df_pp
```

```python
# Filtro 1 (Nombre de vendedor)
filtro1 = df_pp[df_pp["NOMBRE_VENDEDOR"] == "AARON NUÑO CORDOVA"]
filtro1
```

```python
# Convertir filtro 1 en CSV
filtro1.to_csv("filtro1.csv")

# Descargar archivo filtrado en CSV
from google.colab import files
files.download("filtro1.csv")
```

```python
# Filtro 2 (Filtro por fila)
filtro2_pp = df_pp.iloc[3:6, : ]  # Filas consecutivas
filtro2_pp
```

```python
# Convertir filtro 2 en CSV
filtro2_pp.to_csv("filtro2.csv")

# Descargar archivo filtrado en CSV
from google.colab import files
files.download("filtro2.csv")
```

```python
# Filtro 3 (Filtro por columna)
filtro3_pp = df_pp.iloc[:, [0,1,3,8]] #Columnas no consecutivas
filtro3_pp
```

```python
# Convertir filtro 3 en CSV
filtro3_pp.to_csv("filtro3.csv")

# Descargar archivo filtrado en CSV
from google.colab import files
files.download("filtro3.csv")
```
